```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
x = tf.placeholder("float", [None, 784])
W = tf.Variable(tf.zeros([784,10]))
b = tf.Variable(tf.zeros([10]))
```

```python
y = tf.nn.softmax(tf.matmul(x,W) + b)
```

```python
y_ = tf.placeholder("float", [None,10])
```

```python
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for i in range(1000):
        batch_xs, batch_ys = mnist.train.next_batch(100)
        sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
        print("iteration #: ", i, "running")
    s = saver.save(sess,"/tmp/mnst.ckpt")    
    correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print (sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
            
```

```python
def test_slice(i,j):
    a = mnist.train.images[i:j]
    label = mnist.train.labels[i:j]
    check = a.reshape(len(a),784)
    return(check,label)
```

```python
with tf.Session() as sess1:
    saver.restore(sess1,"/tmp/mnst.ckpt")
    check,label = test_slice(0,None)
    ans = sess1.run(y,feed_dict={x:check})
    correct = tf.equal(tf.argmax(ans,1),tf.argmax(label,1))
    accuracy = tf.reduce_mean(tf.cast(correct,'float'))
    print("The Accuracy is:", accuracy.eval())
    #print(sess1.run(tf.argmax(ans,1)))
    #print(sess1.run(tf.argmax(label,1)))    
    
```
